```python
import pandas as pd

#read dataset with pandas and show first 5 rows using .head()
df = pd.read_csv("/Users/farshidayoubzadeh/Desktop/Computer Science/Personal Projects/amazon-orders.csv")
df.head()
```

```python
#chang the NaN values with 0 - part of data cleaning
df = df.fillna(0)
df.head()
```

```python
#change total charged sign with nothing and set it as type float - part of data cleaning
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Total Charged"].sum()
```

```python
df["Total Charged"].mean()
```

```python
df["Total Charged"].median()
```

```python
df["Total Charged"].max()
```

```python
df["Total Charged"].min()
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"].sum()
```

```python
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90)
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10))
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20,10))
```

```python
daily_orders.plot.bar(figsize=(20, 10), color='#61D199')
```

```python
#group by month and sum 
monthly_orders = df.groupby(df['Order Date'].dt.month).sum()['Total Charged']
monthly_orders.head()

```

```python
#plot month spending
graph = monthly_orders.plot.bar( color = '#ffb6c1')
graph.set_xlabel("Order Month")
graph.set_ylabel("Money ($)")
```
